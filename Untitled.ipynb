```python
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# === Paramètres ===
IMG_SIZE = (150, 150)
model_path = "mon_modele_cnn.keras"
model = load_model(model_path)
model.summary()  # utile pour voir la dernière couche convolutive

# === Prétraiter l'image ===
img = load_img(image_path, target_size=IMG_SIZE)
img_array = img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

# === Fonction Grad-CAM ===
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        class_channel = predictions[:, pred_index]

    grads = tape.gradient(class_channel, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    heatmap = np.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

# === Appliquer Grad-CAM ===
last_conv_layer_name = "conv2d_4"  # remplace par ta vraie couche convolutive
heatmap = make_gradcam_heatmap(img_array, model, last_conv_layer_name)

# === Charger et superposer l'image ===
original = cv2.imread(image_path)
original = cv2.resize(original, IMG_SIZE)
heatmap_resized = cv2.resize(heatmap, (original.shape[1], original.shape[0]))
heatmap_colored = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)
superimposed_img = cv2.addWeighted(original, 0.6, heatmap_colored, 0.4, 0)

# === Afficher ===
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(heatmap, cmap='jet')
plt.title("Heatmap")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB))
plt.title("Superimposed")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk
import numpy as np
import cv2
from tensorflow.keras.models import load_model, Model
import matplotlib.pyplot as plt

# Charger le modèle pré-entraîné
saved_model_path = 'mon_modele_cnn.keras'

loaded_model = load_model(saved_model_path)

# Définir le modèle pour extraire les activations d'une couche spécifique
activation_model = Model(inputs=loaded_model.inputs, outputs=loaded_model.layers[3].output)  # Exemple : 4ème couche convolutive

# Variables globales pour l'image chargée et son état
img_pil = None
img_tk = None
colored_img_tk = None

# Fonction pour charger et traiter une image IRM
def load_and_process_image():
    global img_pil, img_tk
    file_path = filedialog.askopenfilename()
    if file_path:
        # Utiliser OpenCV pour charger l'image
        img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)  # Chargement en niveaux de gris
        img = cv2.resize(img, (256, 256))  # Redimensionner l'image selon la taille attendue
       
        # Convertir l'image en format PIL
        img_pil = Image.fromarray(img)
       
        # Convertir l'image PIL en ImageTk pour affichage dans Tkinter
        img_tk = ImageTk.PhotoImage(image=img_pil)
       
        # Affichage de l'image dans l'interface Tkinter
        panel.configure(image=img_tk)
        panel.image = img_tk
       
        # Activer les boutons de classification et de détection
        classify_button.config(state=tk.NORMAL)
        detect_button.config(state=tk.NORMAL)

# Fonction pour segmenter l'image
def segment_image(image):
    # Prétraitement : Filtrage pour réduire le bruit
    image_filtered = cv2.medianBlur(image, 5)
   
    # Appliquer un seuil adaptatif pour une meilleure segmentation
    mask = cv2.adaptiveThreshold(image_filtered, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                 cv2.THRESH_BINARY_INV, 11, 2)
   
    return mask

# Fonction pour appliquer le modèle et colorer les régions affectées
def apply_model_and_color(image):
    # Préparer l'image pour le modèle
    image_resized = cv2.resize(image, (150, 150))  # Adapter à la taille d'entrée du modèle
    image_colored = cv2.cvtColor(image_resized, cv2.COLOR_GRAY2RGB)  # Convertir en image couleur
    image_expanded = np.expand_dims(image_colored, axis=0)
    image_normalized = image_expanded / 255.0
   
    # Obtenir les activations de la couche spécifiée
    activations = activation_model.predict(image_normalized)
   
    # Calculer la moyenne des activations sur l'axe des canaux
    activation_map = np.mean(activations[0], axis=-1)
   
    # Redimensionner les activations à la taille originale de l'image
    activation_map_resized = cv2.resize(activation_map, (256, 256))
   
    # Normaliser les activations
    activation_map_normalized = (activation_map_resized - np.min(activation_map_resized)) / (np.max(activation_map_resized) - np.min(activation_map_resized))
   
    # Intensifier les activations pour les zones critiques
    activation_map_normalized = np.power(activation_map_normalized, 2)
   
    # Convertir en uint8 pour l'affichage
    heatmap = (activation_map_normalized * 255).astype(np.uint8)
   
    # Créer une heatmap rouge
    heatmap_red = np.zeros_like(cv2.cvtColor(heatmap, cv2.COLOR_GRAY2BGR))
    heatmap_red[:, :, 2] = heatmap  # Appliquer les activations au canal rouge uniquement
   
    # Segmenter l'image pour obtenir le masque du cerveau
    mask = segment_image(image)
   
    # Appliquer le masque à la heatmap pour ne colorer que le cerveau
    brain_heatmap = cv2.bitwise_and(heatmap_red, heatmap_red, mask=mask)
   
    # Superposer la heatmap rouge sur l'image originale
    heatmap_superimposed = cv2.addWeighted(brain_heatmap, 0.7, cv2.cvtColor(image, cv2.COLOR_GRAY2BGR), 0.3, 0)
   
    return heatmap_superimposed, activation_map_normalized

# Fonction pour classifier l'image (Non-Demented ou autre)
def classify_image(image):
    image_resized = cv2.resize(image, (150, 150))  # Adapter à la taille d'entrée du modèle
    image_colored = cv2.cvtColor(image_resized, cv2.COLOR_GRAY2RGB)  # Convertir en image couleur
    image_expanded = np.expand_dims(image_colored, axis=0)
    image_normalized = image_expanded / 255.0
   
    prediction = loaded_model.predict(image_normalized)
    predicted_class = np.argmax(prediction, axis=1)
   
    # Afficher la prédiction pour débogage
    print(f"Prédiction du modèle : {prediction}")
    print(f"Classe prédite : {predicted_class}")
   
    # Retourner la classe prédite
    return predicted_class[0]

# Fonction pour classifier et afficher la classe prédite
def classify_and_display():
    global img_pil
   
    if img_pil is None:
        messagebox.showerror("Erreur", "Chargez d'abord une image IRM.")
        return
   
    # Convertir l'image PIL en tableau numpy
    img_array = np.array(img_pil)
   
    # Classifier l'image
    predicted_class = classify_image(img_array)
   
    # Déterminer le nom de la classe
    class_names = ['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented']
    class_name = class_names[predicted_class]
   
    # Afficher la classe prédite
    messagebox.showinfo("Résultat de la Classification", f"La classe prédite est : {class_name}")

# Fonction pour ajouter une barre de couleur sans histogramme et sans ticks ni labels
def add_colorbar(image, activation_map):
    # Convert the image to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
   
    # Create a temporary figure and axis to add the colorbar
    temp_fig, temp_ax = plt.subplots()
    temp_im = temp_ax.imshow(activation_map, cmap='viridis')
    temp_ax.axis('off')
    cbar = temp_fig.colorbar(temp_im, ax=temp_ax, orientation='horizontal', pad=0.04, aspect=50)
    cbar.ax.tick_params(labelsize=0, length=0)  # Hide ticks and labels
   
    # Save the image with the colorbar
    temp_fig.canvas.draw()
    # Convert to ImageTk format
    img_with_colorbar = np.frombuffer(temp_fig.canvas.tostring_rgb(), dtype=np.uint8)
    img_with_colorbar = img_with_colorbar.reshape(temp_fig.canvas.get_width_height()[::-1] + (3,))
   
    plt.close(temp_fig)
   
    # Create the final image without the colorbar
    fig, ax = plt.subplots()
    ax.imshow(image_rgb)
    ax.axis('off')
    fig.canvas.draw()
    img_no_colorbar = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
    img_no_colorbar = img_no_colorbar.reshape(fig.canvas.get_width_height()[::-1] + (3,))
   
    plt.close(fig)
   
    return img_no_colorbar

# Fonction pour détecter les régions affectées et afficher les images
def detect_regions():
    global img_pil, colored_img_tk
   
    if img_pil is None:
        messagebox.showerror("Erreur", "Chargez d'abord une image IRM.")
        return
   
    # Convertir l'image PIL en tableau numpy
    img_array = np.array(img_pil)
   
    # Classifier l'image
    predicted_class = classify_image(img_array)
   
    # Déterminer le nom de la classe
    class_names = ['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented']
    class_name = class_names[predicted_class]
   
    if class_name == 'Non_Demented':
        # Afficher l'IRM sans coloration
        top = tk.Toplevel()
        top.title("Image IRM - Patient Sain")
       
        frame_original = tk.Frame(top)
        frame_original.pack(side=tk.LEFT, padx=10, pady=10)
        label_original = tk.Label(frame_original, text="Image IRM Originale")
        label_original.pack()
        panel_original = tk.Label(frame_original, image=img_tk)
        panel_original.image = img_tk
        panel_original.pack()
       
        frame_message = tk.Frame(top)
        frame_message.pack(side=tk.RIGHT, padx=10, pady=10)
        label_message = tk.Label(frame_message, text="Aucune région affectée détectée.")
        label_message.pack()
       
        return
   
    # Appliquer le modèle et colorer les régions affectées
    colored_image, activation_map = apply_model_and_color(img_array)
   
    # Ajouter une barre de couleur pour interpréter les régions affectées
    colored_image_no_colorbar = add_colorbar(colored_image, activation_map)
   
    # Convertir les images en format ImageTk pour l'affichage dans Tkinter
    colored_img_pil = Image.fromarray(colored_image_no_colorbar)
    colored_img_tk = ImageTk.PhotoImage(image=colored_img_pil)
   
    # Affichage des images originales et de la segmentation colorée
    top = tk.Toplevel()
    top.title("Images détectées")
   
    frame_original = tk.Frame(top)
    frame_original.pack(side=tk.LEFT, padx=10, pady=10)
    label_original = tk.Label(frame_original, text="Image IRM Originale")
    label_original.pack()
    panel_original = tk.Label(frame_original, image=img_tk)
    panel_original.image = img_tk
    panel_original.pack()
   
    frame_colored = tk.Frame(top)
    frame_colored.pack(side=tk.LEFT, padx=10, pady=10)
    label_colored = tk.Label(frame_colored, text="Régions Affectées Colorées")
    label_colored.pack()
    panel_colored = tk.Label(frame_colored, image=colored_img_tk)
    panel_colored.image = colored_img_tk
    panel_colored.pack()

# Interface Tkinter
root = tk.Tk()
root.title("Application IRM")

# Frame pour afficher l'image chargée
frame = tk.Frame(root)
frame.pack(padx=10, pady=10)

# Label pour afficher l'image chargée
panel = tk.Label(frame)
panel.pack()

# Bouton pour charger une image IRM
load_button = tk.Button(root, text="Charger une Image IRM", command=load_and_process_image)
load_button.pack(pady=10)

# Bouton pour classifier l'image IRM
classify_button = tk.Button(root, text="Classifier l'Image", command=classify_and_display, state=tk.DISABLED)
classify_button.pack(pady=10)

# Bouton pour détecter les régions affectées
detect_button = tk.Button(root, text="Détecter les Régions", command=detect_regions, state=tk.DISABLED)
detect_button.pack(pady=10)

root.mainloop()
```
